```python
url= "https://es.wikipedia.org/wiki/Anexo:Provincias_y_ciudades_aut%C3%B3nomas_de_Espa%C3%B1a"
driver= webdriver.Chrome()
driver.get(url)
```

```python
tabla= driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table')
tabla
```

```python
cabecera= tabla.find_element(By.XPATH, 'thead')
```

```python
len(cabecera.find_elements(By.TAG_NAME, 'th'))
```

```python
cabecera= [e.text.replace('\n', ' ').replace('*', '') for e in cabecera.find_elements(By.TAG_NAME, 'th')]
cabecera
```

```python
cuerpo= tabla.find_element(By.TAG_NAME, 'tbody')
filas= cuerpo.find_elements(By.TAG_NAME, 'tr')
len(filas)
```

```python
filas[0].find_elements(By.TAG_NAME, 'td')[0].text
```

```python
matriz=[]
for f in filas:
    tmp=[]
    elemens= f.find_elements(By.TAG_NAME, 'td')
    for e in elemens:
        tmp.append(e.text)
    matriz.append(tmp)
```

```python
len(matriz[0])
```

```python
#Tranformo todo a DF para ordenarlo y representarlo:
columnas=[i for i in range(12)]
```

```python
df= pd.DataFrame(matriz, columns=columnas)
df= df[[2,3,4]]
```
